#### Objective
* To explore how literacy rates impact unemployment trends in India across various districts and states. This analysis will help identify whether higher literacy rates are associated with lower unemployment and provide insights for policymakers to target education-focused interventions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lit = pd.read_csv('Literacy Data 2011.csv')
```

```python
print("Literacy Data:")
print(lit.head())
```

### Dataset Features

**Literacy Data (Census 2011)**
- Features:
* District: Name of the district.
* State: State name corresponding to the district.
* Literacy Rate: Percentage literacy in the district.

```python
lit.info()
```

**Primary Hypothesis: Higher literacy rates correlate with lower unemployment rates in India**

- Key Questions:

1) Do states with higher literacy rates consistently report lower unemployment?
2) Are there notable differences between rural and urban regions in the relationship between literacy and unemployment?
3) Which districts or states deviate significantly from the trend, and why?

```python
literacy_data = lit
```

```python
# Standardizing column names for easier handling
literacy_data.columns = ['S_No', 'District', 'State', 'Literacy']

# Standardizing state names in both datasets (if inconsistencies exist)
literacy_data['State'] = literacy_data['State'].str.strip().str.upper()
```

```python
literacy_data
```

#### Top 50 districts with the highest literacy

```python
# Top 50 districts with the highest literacy
top_literacy = literacy_data.nsmallest(50, 'Literacy')

# Bar chart for top districts by literacy
plt.figure(figsize=(10, 6))
sns.barplot(data=top_literacy, x='Literacy', y='District', palette='viridis')
plt.title('Top 50 Districts by Illiteracy Rate')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('District')
plt.show()
```

#### Boxplot to observe literacy rate distribution across states

```python
# Boxplot to observe literacy rate distribution across states
plt.figure(figsize=(12, 6))
sns.boxplot(data=literacy_data, x='State', y='Literacy')
plt.xticks(rotation=90)
plt.title('Box Plot of Literacy Rate Distribution by State')
plt.ylabel('Literacy Rate (%)')
plt.xlabel('State')
plt.show()
```

#### Rank of the states based on the literacy rates

```python
state_literacy_order = literacy_data.groupby('State', as_index=False)['Literacy'].mean()
state_literacy_order = state_literacy_order.sort_values(by='Literacy', ascending=False)

# Order of states based on literacy rate
print("Order of States by Literacy Rate (Descending):")
print(state_literacy_order)
```

```python
from matplotlib.cm import get_cmap

# Grouping the data by state and calculating average literacy rate
state_literacy_order = literacy_data.groupby('State', as_index=False)['Literacy'].mean()
# Sorting the states
state_literacy_order = state_literacy_order.sort_values(by='Literacy', ascending=False)

viridis = get_cmap('viridis')
colors = viridis(state_literacy_order['Literacy'] / max(state_literacy_order['Literacy']))

plt.figure(figsize=(12, 8))
plt.bar(state_literacy_order['State'], state_literacy_order['Literacy'], color=colors)

plt.title('Average Literacy Rate by State (Descending Order)', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Average Literacy Rate (%)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

#### Classifying the literacy rate of the states in the 3 categories using KMeans

```python
from sklearn.cluster import KMeans
literacy_rates = literacy_data[['Literacy']]

# Normalizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
literacy_rates_scaled = scaler.fit_transform(literacy_rates)

# Applying K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=986)
literacy_data['Cluster'] = kmeans.fit_predict(literacy_rates_scaled)
```

```python
import plotly.express as px
fig = px.scatter(
    literacy_data,
    x="District",
    y="Literacy",
    color="Cluster",
    title="Clustering Districts by Literacy Rate",
    labels={"District": "District", "Literacy": "Literacy Rate"},
    color_discrete_sequence=px.colors.qualitative.Set1
)
fig.update_layout(
    xaxis_title="District",
    yaxis_title="Literacy Rate",
    legend_title="Cluster",
    width=800,
    height=600
)
fig.show()
```
