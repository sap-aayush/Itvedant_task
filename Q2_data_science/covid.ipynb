```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("day_wise.csv", parse_dates=['Date'])
```

#### Confirmed Cases Over Time

```python
# Confirmed Cases over time
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Confirmed'], color='blue', label='Confirmed Cases')
plt.title('Confirmed Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Deaths, Recovered, and Active Cases Over Time

```python
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Deaths'], label='Deaths', color='red')
plt.plot(df['Date'], df['Recovered'], label='Recovered', color='green')
plt.plot(df['Date'], df['Active'], label='Active Cases', color='orange')
plt.title('Deaths, Recovered, and Active Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Bar plot of New Cases, New Deaths, and New Recovered by Date

```python
plt.figure(figsize=(10,6))
plt.bar(df['Date'], df['New cases'], label='New Cases', color='blue', alpha=0.6)
plt.bar(df['Date'], df['New deaths'], label='New Deaths', color='red', alpha=0.6)
plt.bar(df['Date'], df['New recovered'], label='New Recovered', color='green', alpha=0.6)
plt.title('New Cases, New Deaths, and New Recovered by Date')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Deaths / 100 Cases

```python
plt.figure(figsize=(10,6))
plt.bar(df['Date'], df['Deaths / 100 Cases'], label='Deaths / 100 Cases', color='red')
plt.title('Deaths / 100 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Deaths / 100 Cases')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Line plot of Recovered / 100 Cases vs Deaths / 100 Recovered Over Time

```python
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Recovered / 100 Cases'], label='Recovered / 100 Cases', color='green')
plt.plot(df['Date'], df['Deaths / 100 Recovered'], label='Deaths / 100 Recovered', color='red')
plt.title('Recovered / 100 Cases vs Deaths / 100 Recovered Over Time')
plt.xlabel('Date')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
correlation_df = df[['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths', 'New recovered', 'Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered']]
corr = correlation_df.corr()

# Plotting the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
country_counts = df['No. of countries'].value_counts()

plt.figure(figsize=(10, 10))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set3", len(country_counts)))
plt.title('Distribution of No. of Countries Over Time')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.stackplot(df['Date'], df['Confirmed'], df['Deaths'], df['Recovered'], labels=['Confirmed', 'Deaths', 'Recovered'], alpha=0.7)
plt.title('Stacked Area Chart: Confirmed, Deaths, and Recovered Cases')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
