```python
import pandas as pd
df = df.drop(columns=["disability"])

categorical_columns = [column for column in df.columns if df[column].dtype == 'object' and not df[column].str.isnumeric().all()]
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=False)

print(df_encoded.head())
```

```python
# Remove columns that are completely empty (all NaN values)
df_cleaned = df_encoded.dropna(axis=1, how='all')

# If you want to drop columns that have a high percentage of missing values, you can set a threshold
# Example: Drop columns with more than 30% missing values
threshold = 0.3
df_cleaned = df_encoded[df_encoded.columns[df_encoded.isnull().mean() < threshold]]
```
